### derivatives

```python
def deravitive_x(f,x,step_size):
    m=(f(x+step_size)-f(x))/((x+step_size)-x)
    return m
def my_function(x):
    return x**2
slope_at_2=deravitive_x(my_function,2,0.00001)
slope_at_2
```
