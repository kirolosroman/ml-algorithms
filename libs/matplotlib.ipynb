## matplotlib example workflow

```python
# 1-import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 2-prepare data
x=[1,2,3,4,5]
y=[11,22,33,44,55]

# 3-setup plot
fig,ax=plt.subplots(figsize=(10,5))

#4-plot the data
ax.plot(x,y)

#5-customize the plot
ax.set(title='simple plot',xlabel="X-axis",ylabel="Y-axis")

#6- save and show

fig.savefig("C:\krh\dev\python\ml\plots\sampleplot.png")
```

## plotting more than one subplot

```python
# two options for subplots

# option 1
x=np.random.random(20)
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.bar(x,x*2)
ax2.scatter(x,np.exp(x))
ax3.plot(x,np.sin(x))
ax4.plot(x,np.cos(x))
```

```python
# option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].bar(x,x*2)
ax[0,1].scatter(x,np.exp(x))
ax[1,0].plot(x,np.sin(x))
ax[1,1].plot(x,np.cos(x))
```

```python
df=pd.read_csv("C:\krh\dev\python\ml\datsets\heart-disease.csv")
df
```

```python
df.head()
```

```python
over_50=df[df['age']>50]
over_50
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,5))
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
ax.set_xlim([50,75])
ax.set_ylim([150,450])
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'], y=over_50['chol'],c=over_50['target'])

ax.set(title="heart disease and chol levels"
       ,xlabel="Age",ylabel="Chol levels")

ax.legend(*scatter.legend_elements(),title="target")

ax.axhline(over_50['chol'].mean(),linestyle="--")
```
