## 1- getting the data ready

```python
import pandas as pd
import numpy as np

heart_disease=pd.read_csv("C:\krh\dev\python\ml\datsets\heart-disease.csv")
heart_disease
```

### create X features Matrix & Y labels

```python
X=heart_disease.drop("target",axis=1)
Y=heart_disease["target"]
```

## 2- choose the right model and hyperparameters
### in this case ..the problem we have is a classification problem

### In this problem we will use Random Forest Classifier 
- Random Forests is an ensemble learning method for classification and regression and other tasks that operates by constructing multitude of decision trees at training time.
  so a collection of modeles will be used to make predictions rather than individual model and this will the overall performance.



### Main ensemble methods in machine learning
### 1- bagging
if we have a dataset and we make different models on this dataset then there is a chance that all these models will have the same output because of the same input data.
so to avoid this mistake we use a technique called BOOTSTRAPING.
bootstrapping: create subsets of the original dataset with replacement.
the size of the subsets is the same as the size of the original dataset.


### 2- boosting
boosting technique is a sequential process, where each model tries to correct the errors of the previous model. the successeding models are dependent on the previous model.

### Gini Index 
simply we need to know the purity and impurity of the dataset so we make sure the forrest increased in performance not the opposite using Gini Index it will help us to differentiate between the impurity of the decision nodes 

### Gini Index 
### = $1- \sum_{i=1}^n (p_i)^2 $  
### = $1- [(p_+)^2+(p_-)^2]$

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()
clf.get_params()
```

```python
## 3- fit the model to the training data
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
clf.fit(X_train,Y_train)
```

## 4- making predictions with the chosen model

```python
y_preds=clf.predict(X_test)
y_preds
```

## evaluate the model on the training data and test data

```python
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

### classification report

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print(classification_report(Y_test,y_preds))
```

### confusion matrix

```python
confusion_matrix(Y_test,y_preds)
```

### accuracy_score

```python
accuracy_score(Y_test,y_preds)
```

## 6-improving the model predictions
### by trying different amounts of n_estimators

```python
np.random.seed(42)
for i in range(10,100,10):
    print(f"trying model with {i} estimators")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train, Y_train)
    print(f"model accuracy: {clf.score(X_test,Y_test)*100:.2f}")
```

## 7- save the model and load it

```python
import pickle
pickle.dump(clf,open("random_forest_model_heartd.pkl","wb"))
```
